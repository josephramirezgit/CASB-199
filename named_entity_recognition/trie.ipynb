```python
# Import the relevant txt file
with open ('../classification/data/protein_aliases.txt') as f:
    lines = f.readlines()


# Create a dictionary to be sent into the trie
input_synonyms = dict()
for line in lines:
    items = line.split('|')
    uid = items[0]
    synonyms = items[1:]
    for synonym in synonyms:
        input_synonyms[synonym.replace("_", " ").lower().strip()] = uid
```

```python
prefix_tree = pygtrie.CharTrie(input_synonyms)
```
